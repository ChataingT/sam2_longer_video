```python
import decord
from decord import VideoReader, cpu
import matplotlib.pyplot as plt

# Load the video
video_path = r'D:\idtracking\data\video\7965_T2a_ADOS_freeplay.mp4'
vr = VideoReader(video_path, ctx=cpu(0))

# Read the first frame
frame = vr[0].asnumpy()

# Save the first frame as an image
plt.imsave('first_frame.png', frame)

# Display the first frame
plt.imshow(frame)
plt.title('First Frame')
plt.show()

# Example: Get the color of the pixel at (x, y)
x, y = 100, 150  # Change these coordinates as needed
pixel_color = frame[y, x]
print(f'The color of the pixel at ({x}, {y}) is {pixel_color}')
```
